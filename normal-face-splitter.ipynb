```python
import os
import shutil

BASEPATH = os.path.abspath(os.getcwd())
for dir_person in os.listdir(BASEPATH+"\Strokefaces\lfw_funneled"):
    if os.path.isdir(BASEPATH+"\lfw_funneled\\"+dir_person):
        for dir_face in os.listdir(BASEPATH+"\lfw_funneled\\"+dir_person):
            #get only 1 photo per person therefore delete a break is needed
            if dir_face.endswith(".jpg"):
                shutil.copy(BASEPATH+"\lfw_funneled\\"+dir_person+"\\"+dir_face, BASEPATH+"\\"+"normal"+"\\"+dir_face)
                print("coppied : "+dir_face)
                break
```

```python
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray
from mtcnn.mtcnn import MTCNN

# extract a single face from a given photograph
def extract_face(filename, required_size=(224, 224)):
	# load image from file
	pixels = plt.imread(filename)
	# create the detector, using default weights
	detector = MTCNN()
	# detect faces in the image
	results = detector.detect_faces(pixels)
	# extract the bounding box from the first face
	x1, y1, width, height = results[0]['box']
	x2, y2 = x1 + width, y1 + height
	# extract the face
	face = pixels[y1:y2, x1:x2]
	# resize pixels to the model size
	image = Image.fromarray(face)
	image = image.resize(required_size)
	return image
```

```python
for photo in os.listdir(BASEPATH+"\Strokefaces\droopy"):
    try:
        # load the photo and extract the face
        image = extract_face(BASEPATH+"\Strokefaces\droopy\\"+photo)
        # save the face in a new file
        image.save(BASEPATH+"\Strokefaces\droopy\\"+photo)
    except:
        pass

for photo in os.listdir(BASEPATH+"\Strokefaces\\normal"):
    try:
        # load the photo and extract the face
        face_extract = extract_face(BASEPATH+"\Strokefaces\\normal\\"+photo)
        face_extract.save(BASEPATH+'\stroke-normal_dataset_processed\\normal\\'+photo)
    except:
        pass
```
